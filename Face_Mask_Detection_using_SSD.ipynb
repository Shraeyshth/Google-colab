```python
train_generator = train_datagen.flow_from_directory(
    'observations-master',
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical"  # This will ensure the labels are one-hot encoded
)
```

```python
model = build_ssd_model()

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=10
)
```

```python
import cv2
import numpy as np

# Function to detect mask in a frame
def detect_mask(frame, model):
    image = cv2.resize(frame, (224, 224))
    image = image.astype("float") / 255.0
    image = np.expand_dims(image, axis=0)

    (mask, withoutMask) = model.predict(image)[0]
    label = "Mask" if mask > withoutMask else "No Mask"

    return label

# Real-time webcam capture and mask detection
cap = cv2.VideoCapture(0)  # Use 0 to capture from webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    label = detect_mask(frame, model)
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
    cv2.imshow("Face Mask Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
