```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing(as_frame=True)
df = pd.DataFrame(data=housing.data, columns=housing.feature_names)
df['target'] = housing.target
```

```python
corr_matrix = df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for California Housing Dataset")
plt.show()
```

```python
housing = fetch_california_housing(as_frame=True)
df = pd.DataFrame(data=housing.data, columns=housing.feature_names)
```

```python
sns.kdeplot(data=df, x="MedInc", fill=True)
plt.title("KDE Plot of Median Income")
plt.show()
```

```python
sns.kdeplot(data=df, x="MedInc", y="AveRooms", fill=True)
plt.title("Bivariate KDE Plot of Median Income vs. Average Rooms")
plt.show()
```
